```python
# Paso 2: Importar librerías
import pygad
import numpy as np
import matplotlib.pyplot as plt

# Paso 3: Definir la función de fitness con 3 parámetros según PyGAD 2.20.0
def fitness_func(ga_instance, solution, solution_idx):
    x = solution[0]
    fitness = x ** 2
    return fitness

# Paso 4: Definir el espacio genético (rango de valores para x)
gene_space = {'low': -10, 'high': 10}

# Paso 5: Configurar el algoritmo genético
ga_instance = pygad.GA(
    num_generations=30,
    num_parents_mating=4,
    fitness_func=fitness_func,
    sol_per_pop=10,
    num_genes=1,
    gene_space=gene_space,
    parent_selection_type="tournament",
    crossover_type="single_point",
    mutation_type="random",
    mutation_percent_genes=100
)

# Paso 6: Ejecutar el algoritmo
ga_instance.run()

# Paso 7: Mostrar la mejor solución
solution, solution_fitness, solution_idx = ga_instance.best_solution()
print(f"Mejor x: {solution[0]:.4f}")
print(f"Fitness (x^2): {solution_fitness:.4f}")

# Paso 8: Graficar la evolución del fitness
ga_instance.plot_fitness()
```
